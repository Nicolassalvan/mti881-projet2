```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
import numpy as np
path_predictions = 'local1/predictions.csv'
df = pd.read_csv(path_predictions, sep=',', header=0)
confusion_matrix = pd.crosstab(df['y_true'], df['y_pred']) 
confusion_matrix = confusion_matrix / confusion_matrix.sum() * 100  # Normalize to percentage
x, y = confusion_matrix.shape 
x, y = x/2, y/2 

fig, ax = plt.subplots(figsize=(y, x), dpi=300) # MODIFIER ICI
sns.heatmap(confusion_matrix, annot=True, cmap='Blues', fmt='.0f', ax=ax)
ax.set_xlabel('Predicted Label')
ax.set_ylabel('True Label')
ax.set_title('Confusion Matrix')
ax.set_yticklabels(ax.get_yticklabels(), rotation=0)
fig.savefig('confusion_matrix.png', dpi=300, bbox_inches='tight') # MODIFIER ICI
plt.show()
```
